```python
# E) Ufuncs y where
v = np.linspace(-2, 2, 5)   # [-2,-1,0,1,2]
sq = np.sqrt(np.abs(v))     # raíz de positivos
print(v)
w = np.array([10,20,30,40,50])
sel = np.where(v >= 0, w, 0)  # si v>=0 toma w, si no 0
sq, sel
```

```python
# Crea S con enteros aleatorios en [0, 100) y convierte a float32

s = np.random.uniform(0, 100, size=(100,4)).astype(np.float32)
s
```

```python
# Centra cada columna restando su media y escala por su desviación típica (z-score)

media_col = np.mean(s, axis=0)
media_fil = np.mean(s, axis=1)
std_col = np.std(s, axis=0, ddof=1)

z = (s - media_col) / (std_col)
z
```

```python
# Construye la matriz de covarianzas 4×4 usando solo operaciones de NumPy (sin np.cov), partiendo de los datos z-score.

n = s.shape[0]
zt = z.T

cov = (zt @ z) / (n - 1)
cov, n

```

```python
# Calcula el producto G = Z.T @ Z / (n-1) y verifica que coincide con tu covarianza
G = z.T @ z / (n-1)
G
```

```python
# Usa where para crear una máscara booleana que marque días “anómalos” si algún sensor z-score |z|≥2 (pista: np.any por filas).

anomalous_mask = np.any(np.abs(z) >= 2, axis = 1)

indices = np.where(anomalous_mask)[0]

```

```python
# Extra: añade a cada fila un vector patrón [0.5, -0.5, 0, 1] via broadcasting y vuelve a detectar anómalos. Explica si cambia algo.
# 1) vector patrón (asegúrate del tipo)
pattern = np.array([0.5, -0.5, 0.0, 1.0], dtype=s.dtype)  # shape (4,)

# 2) suma por broadcasting: s tiene shape (100,4)
s2 = s + pattern   # shape (100,4) -- broadcasting por filas

# 3) recalcula z-scores (igual que antes)
mean2 = np.mean(s2, axis=0)
std2  = np.std(s2, axis=0)           # usa ddof=0 si antes lo usaste así
z2    = (s2 - mean2) / std2

# 4) detecta anomalías de nuevo
anomalous_mask2   = np.any(np.abs(z2) >= 2, axis=1)
indices2          = np.where(anomalous_mask2)[0]

# 5) compara con la máscara/índices anteriores
print("Anomalous original:", anomalous_mask.sum(), "days")
print("Anomalous after pattern:", anomalous_mask2.sum(), "days")
print("Masks equal (allclose/bool equal):", np.array_equal(anomalous_mask, anomalous_mask2))
print("Indices equal (first 20):", np.array_equal(indices[:20], indices2[:20]))
```
